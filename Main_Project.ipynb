### <b><span style='color:coralred'> 1.3. Bussiness Task</span></b>

**Bagaimana Cara Menurunkan Nilai Recency Customer?**
Recency dengan Nilai lebih dari 30, menandakan pelanggan tidak kembali pada 1 bulan terkahir (berlaku kelipatannya) dan itu menjadi indikator bahwa kesehatan mengelola pelanggan yang kurang ideal. Tingkat `Churn` ideal untuk perusahaan yang matang dan mapan adalah 5% hingga 7% dalam churn tahunan dan kurang dari 1% dalam churn bulanan. Jika perusahaan memiliki 1.000 pelanggan, ini berarti kita hanya akan kehilangan 50 pelanggan per tahun atau empat hingga lima Customer per bulan. 

Oleh karena itu, diperlukan strategi yang komprehensif untuk mengidentifikasi dan menangani faktor-faktor yang mempengaruhi `Recency`.
Ada 4 hal utama yang menjadi fokus pada hubungan antara `Nilai Recency` yang rendah dengan faktor lainya:

#### 1. Customer Demographics
* Bagaimana hubungan usia customer dengan `Recency`?
* Bagaimana pengaruh tingkat pendidikan terhadap `Recency`?

#### 2. Customer Income and Spending
* Bagaimana distribusi pendapatan di antara customer dengan `Recency` rendah?
* Apakah ada korelasi antara pendapatan dan total pembelian produk (MntWines, MntFruits, MntMeatProducts, dll) dengan `Recency`?

#### 3. Purchase Behavior
* Bagaimana frekuensi pembelian melalui website mempengaruhi `Recency`?
* Bagaimana pola pembelian melalui katalog dan toko offline terkait dengan `Recency`?

#### 4. Promotion Response
* Apakah kampanye promosi efektif dalam menurunkan `Recency`?
* Bagaimana efektivitas diskon dalam menurunkan `Recency`?

Analisis ini akan fokus pada Customer Demographic, Purchase Behaviour, Engagement Metrics (Promotion Response). Pelanggan dengan keterlibatan lebih dari 30 hari (1 Bulan) akan diprioritaskan untuk mencerminkan pola interaksi yang berarti, yang artinya pada analisis ini akan menaruh batas (threshold) pada angka 30 untuk mencerminkan standar `Churn` potential diatas angka tersebut. Secara khusus, analisis ini bertujuan untuk mengidentifikasi pain point dalam faktor `Recency` yang mempengaruhi loyalitas dan **mencapai pengurangan 5% dalam churn selama tahun fiskal berikutnya**.

##### <b><span style='color:coralred'> 2.2.2. Data Condition: Quality and Representativeness</span></b>

```python
from scipy.stats import shapiro

# variabel berisi column
columns = ['ID', 'Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency',
           'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
           'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
           'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
           'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
           'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Z_CostContact',
           'Z_Revenue', 'Response']

# Membuat grid subplot
fig, axes = plt.subplots(6, 5, figsize=(20, 15))

# Meratakan array axes untuk iterasi yang mudah
axes = axes.flatten()

# Melakukan uji Shapiro-Wilk untuk setiap kolom dan membuat histogram
for i, column in enumerate(columns):
    ax = axes[i]
    # Cek apakah kolom adalah numerik; jika tidak, skip uji Shapiro-Wilk
    if pd.api.types.is_numeric_dtype(df[column]):
        sns.histplot(data=df, x=column, kde=True, label=column, ax=ax)
        ax.set_title(column)

        # Melakukan uji Shapiro-Wilk
        stat, p = shapiro(df[column])

        if p > 0.05:
            ax.set_xlabel(f"Shapiro-Wilk Test: p={p:.3f}\nNormal Distribution")
        else:
            ax.set_xlabel(f"Shapiro-Wilk Test: p={p:.3f}\nNot Normal Distribution")
    else:
        ax.set_xlabel("Non-numeric data")

# Menghapus subplot yang tidak digunakan
for i in range(len(columns), len(axes)):
    fig.delaxes(axes[i])

# Menyesuaikan tata letak dan menampilkan plot
plt.tight_layout()
plt.show()
```

```python
def winsorize(data, lower_percentile=0.10, upper_percentile=0.90):
    lower_limit = data.quantile(lower_percentile)
    upper_limit = data.quantile(upper_percentile)
    data_cleaned = data.clip(lower=lower_limit, upper=upper_limit)
    return data_cleaned

AWines = winsorize(df['MntWines'])
df['MntWines'] = AWines

AFruits = winsorize(df['MntFruits'])
df['MntFruits'] = AFruits

AMeat = winsorize(df['MntMeatProducts'])
df['MntMeatProducts'] = AMeat

AFish =winsorize(df['MntFishProducts'])
df['MntFishProducts'] = AFish

ASweet = winsorize(df['MntSweetProducts'])
df['MntSweetProducts'] = ASweet

AGold = winsorize(df['MntGoldProds'])
df['MntGoldProds'] = AGold
```

```python
# Menambah Kolom 'mnttotal dari kolom 'MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds
df.insert(19, 'mnttotal', df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'])
```

```python
# Menambah Kolom 'totalpurchases dari kolom NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases
df.insert(25, 'totalpurchases', df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'])
```

```python
# Menambah Kolom 'totalacceptedcmp dari kolom NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases
df.insert(31, 'totalacceptedcmp', df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response'])
```
